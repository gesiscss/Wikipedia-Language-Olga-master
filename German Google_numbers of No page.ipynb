```python
import json
%matplotlib inline
import collections
import matplotlib.pyplot as pyplot
from matplotlib.legend_handler import HandlerLine2D
import operator

def load_simple_json(filename):
    with open(filename, 'r') as f:
        return json.load(f)

m_no_page=load_simple_json('de/wiki/m_no_page.json')
f_no_page=load_simple_json('de/wiki/f_no_page.json')
m_page=load_simple_json('de/wiki/m_page.json')
f_page=load_simple_json('de/wiki/f_page.json')
m_page_o=load_simple_json('de/wiki/m_page_other.json')
m_links_to_feminine=load_simple_json('de/wiki/m_links_to_feminine.json')
m_links_to_smth=load_simple_json('de/wiki/m_links_to_smth.json')
f_links_to_mascuilne=load_simple_json('de/wiki/f_links_to_mascuilne.json')
f_links_to_smth=load_simple_json('de/wiki/f_links_to_smth.json')
words=load_simple_json('de/occupation_all.json')#{masculine:[[],[feminine]]}
words_revert={words[i][1]:i for i in words} #feminine:masculine
#beruf_subcategories=load_simple_json('de/wiki/beruf_subcategories.json')

google_number=load_simple_json('de/googlenumber_new2.json')

y_val=[]
#look what are the google scores of m_no_page
for i in m_no_page:
    if google_number.has_key(i):
        y_val.append(int(google_number[i][0]))  

y_val_f=[]
#look what are the google scores of f_no_page
for i in f_no_page:
    if words_revert.has_key(i):
        if google_number.has_key(words_revert[i]):
            y_val_f.append(int(google_number[words_revert[i]][1]))  
y_val=sorted(y_val)
y_val_f=sorted(y_val_f)
pyplot.xlabel('google results')
pyplot.ylabel('occupations ranked')
pyplot.title('Distribution of google results of occupations without wiki page(German)')
line1,=pyplot.loglog(range(1,len(y_val)+1),y_val,'o',ms=5,label="masculine")
line2,=pyplot.loglog(range(1,len(y_val_f)+1),y_val_f,'o',ms=5,label="feminine")

pyplot.legend(loc=2, handler_map={line1: HandlerLine2D(numpoints=4),line2: HandlerLine2D(numpoints=4)})
```
