```python
#store pages from category Beruf




from wikitools import wiki, api,category
import json

def seen_wiki_categories(lang,category_page,overloop=0,n=6):
    global seen_categories_
    global seen_pages_
    where_go_categories_=[]
    #print "We are here: "+ category_page
    seen_categories_.append(category_page)
    site = wiki.Wiki("http://"+lang+".wikipedia.org/w/api.php") 
    cat = category.Category(site, category_page) # Create object for "Category:xxxxx"
    for article in cat.getAllMembersGen(titleonly=True,namespaces=[0]): # iterate through all the pages in ns 0
        if article not in seen_pages_ :
            if "," not in article:
                seen_pages_.append(article)
        else:
            continue
    for subcategory in cat.getAllMembersGen(titleonly=True,namespaces=[14]):
        if subcategory not in seen_categories_ :
            where_go_categories_.append(subcategory)
        else:
            continue
    for c in where_go_categories_ :
        if (c not in seen_categories_)&(c!="Категория:Люди по профессиям".decode('utf8'))&\
        (c!="Категория:Профессиональные заболевания".decode('utf8'))&\
        (c!="Категория:Профессиональные праздники".decode('utf8'))&\
        (c!="Категория:Водолазное дело".decode('utf8'))&\
        (c!="Категория:Профессиональные династии".decode('utf8'))&\
        (c!="Категория:Таксидермия".decode('utf8'))&(c!="Категория:Оннагата".decode('utf8'))&\
        (c!="Категория:Типографы".decode('utf8'))&(c!="Категория:Уличные газеты".decode('utf8'))&\
        (c!="Категория:Сварщики".decode('utf8'))&(c!="Категория:Токари".decode('utf8'))&\
        (c!="Категория:Фрезеровщики".decode('utf8'))&(c!="Категория:Кузнечное ремесло".decode('utf8'))&\
        (c!="Категория:Гусляры".decode('utf8'))&(c!="Категория:Дворецкие".decode('utf8'))&\
        (c!="Категория:Шуты".decode('utf8'))&(c!="Портал:Профессии".decode('utf8')):
            if overloop<n:
                seen_wiki_categories(lang,c,overloop=overloop+1,n=n)
            else:
                print "overloop ==="+c
                pass
                     


            
"""   
[×] Косторезы‎ (3: 3 с.)

Военные‎ (36: 29 кат., 7 с.)
    ► Военные инженеры‎ (74: 2 кат., 72 с.)
    ► Снайперы‎ (6: 3 кат., 3 с.)
    
 *
    Ведущий программист
    Программист   
    
     Программисты‎ (9: 2 кат., 7 с.)
     
    [×] Авторы книг по кулинарии‎ (46: 46 с.)
    [×] Дегустаторы‎ (1: 1 с.)
    ► Кондитеры‎ (2: 1 кат., 1 с.)
    ► Повара‎ (5: 2 кат., 3 с.)
    ► Рестораторы‎ (5: 1 кат., 4 с.)
    ► Рыбаки‎ (6: 1 кат., 5 с.)
    
  Ведущие по алфавиту‎ (4: 4 с.)
► Кинематографисты‎ (39: 27 кат., 12 с.)
► Медиамагнаты‎ (35: 1 кат., 34 с.)
► Редактор‎ (2: 1 кат., 1 с.)
[×] Репортёры‎ (39: 39 с.) 
  
 [×] Акушерки‎ (3: 3 с.)
 
 
    [×] Аранжировщики‎ (24: 24 с.)
    [×] Гусляры‎ (6: 6 с.)
    ► Диджеинг‎ (15: 2 кат., 13 с.)
    ► Дирижёры‎ (17: 15 кат., 2 с.)
    ► Поэты-песенники‎ (28: 2 кат., 26 с.)
 
 ► Адвокаты‎ (10: 4 кат., 6 с.)
► Азартные игры как профессия‎ (8: 1 кат., 7 с.)

► Гейша‎ (33: 3 кат., 30 с.)
► Няни‎ (14: 3 кат., 11 с.)
► Рестораторы‎ (5: 1 кат., 4 с.)
  
    """
l="ru"
seen_categories_=[]
seen_pages_=[]
seen_wiki_categories(l,"Категория:Профессии",overloop=0,n=2)
seen_pages_=list(set(seen_pages_))
print len(seen_pages_) #632
with open(l+'/wiki/berufe_wiki.json', 'w') as f: 
    json.dump(seen_pages_,  f,  indent=4,ensure_ascii=False)
```

```python

def check_pages_in_list(both6):
    new_both6={}
    for i in both6:
        if i not in new_both6.keys():
            new_both6[i]=both6[i]
        for j in both6[i]:
            if j in m_page:
                index=int(both6[i].index(j))
                new_both6[i][index]=both6[i][index],1 #1=page exist
            elif j in m_redirection:
                index=int(both6[i].index(j))
                new_both6[i][index]=both6[i][index],3 #3=redirects to smth
            elif j in m_redirection.values():
                index=int(both6[i].index(j))
                new_both6[i][index]=both6[i][index],4 #4=redirects to smth value
            else:
                index=int(both6[i].index(j))
                new_both6[i][index]=both6[i][index],0 #no page
    return new_both6
new_both6=check_pages_in_list(both6)
```
